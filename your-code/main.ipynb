## Regression Model Evaluation

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('/housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
"""
CRIM - per capita crime rate by town
ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
INDUS - proportion of non-retail business acres per town.
CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX - nitric oxides concentration (parts per 10 million)
RM - average number of rooms per dwelling
AGE - proportion of owner-occupied units built prior to 1940
DIS - weighted distances to five Boston employment centres
RAD - index of accessibility to radial highways
TAX - full-value property-tax rate per $10,000
PTRATIO - pupil-teacher ratio by town
B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
LSTAT - % lower status of the population
MEDV - Median value of owner-occupied homes in $1000's"""
```

```python
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
X = data.drop("MEDV", axis = 1)
y = data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size=0.2)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

## 3. Calculate and print R-squared for both the training and the testing set.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
#mean squared error for training data set
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, model.predict(X_train))
print(f"mean squared error for training data set: {mean_squared_error(y_train, model.predict(X_train))}")

#mean squared error for the test data set
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
print(f"mean squared error for test data set: {mean_squared_error(y_test, y_pred)}")
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
#absolute error for the training data set
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_train, model.predict(X_train))
print(f"mean absolute error for training data set: {mean_absolute_error(y_train, model.predict(X_train))}")

#absolute error for the test data set
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
print(f"mean absolute error for test data set: {mean_absolute_error(y_test, y_pred)}")
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
#select data into X and y before performing a train_test_split
X = df
y = pd.DataFrame(data.target)
```

```python
#perform a train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size=0.2)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
#Use a Logisticregression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
#Test data accuracy
print(f"test data accuracy: {model.score(X_test,y_test)}")

#train data accuracy
print(f"train data accuracy: {model.score(X_train,y_train)}")
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
#calculate precision score for the training data set
from sklearn.metrics import precision_score
precision_score(y_train, model.predict(X_train), average='micro')
print(f"precision score for the training data set: {precision_score(y_train, model.predict(X_train), average='micro')}")

#calculate precision score for the testing data set
from sklearn.metrics import precision_score
precision_score(y_test, model.predict(X_test), average='micro')
print(f"precision score for the testing data set: {precision_score(y_test, model.predict(X_test), average='micro')}")
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
#calculate recall score for the training data set
from sklearn.metrics import recall_score
recall_score(y_train, model.predict(X_train), average='micro')
print(f"recall score for the training data set: {recall_score(y_train, model.predict(X_train), average='micro')}")

#calculate recall score for the testing data set
from sklearn.metrics import recall_score
recall_score(y_test, model.predict(X_test), average='micro')
print(f"recall score for the testing data set: {recall_score(y_test, model.predict(X_test), average='micro')}")
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
#calculate F1 score for the training data set
from sklearn.metrics import f1_score
f1_score(y_train, model.predict(X_train), average='micro')
print(f"F1 score for the training data set: {f1_score(y_train, model.predict(X_train), average='micro')}")

#calculate F1 score for the testing data set
from sklearn.metrics import f1_score
f1_score(y_test, model.predict(X_test), average='micro')
print(f"F1 score for the testing data set: {f1_score(y_test, model.predict(X_test), average='micro')}")
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
#Generate confusion matrices for the training data set
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, model.predict(X_train))
print(f"confusion matrix for the training data set: {confusion_matrix(y_train, model.predict(X_train))}")

#generate confusion metrics for the test data set
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, model.predict(X_test))
print(f"confusion matrix for the test data set: {confusion_matrix(y_test, model.predict(X_test))}")
```
